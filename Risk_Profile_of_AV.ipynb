**Risk Profile**

```python
import matplotlib.pyplot as plt

# Data for the risk profile
profits = [78.75, 157.5]
probabilities = [0.4, 0.6]

# Create the plot
plt.figure(figsize=(8, 6))
plt.stem(profits, probabilities, basefmt=" ", use_line_collection=True)

# Adding labels and title
plt.xlabel('Profit (Million USD)')
plt.ylabel('Probability')
plt.title('Risk Profile')

# Show the plot
plt.grid(True)
plt.show()
```
